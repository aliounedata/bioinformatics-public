# Purpose: Visualization

### In this Notebook we will visualize the pIC50 for COVID-19 molecules.

The data in this project was inspired by [Chanin Nantasenamat (PhD)](https://www.linkedin.com/in/chanin-nantasenamat/) on his [YouTube](https://www.youtube.com/c/DataProfessor/about) Tutorials.

```python
import pandas as pd
import mols2grid
```

```python
df = pd.read_csv('processed.csv')
df.head()
```

```python
mols2grid.display(df,
                  fixedBondLength=25,
                  rename={"canonical_smiles": "SMILES", "molecule_chembl_id": "Name"},
                  subset=["Name","img","pIC50"],
                  tooltip=["Name","pIC50"],
                  transform={"pIC50": lambda x: f"{x:+.2f}"})
```
